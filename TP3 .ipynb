#### 5. Un router recibe un datagrama con header básico y longitud total 832 bytes.  Debe  ser  enviado  a  un  segmento  en  el  que  el  MTU  es  de  520 bytes. Determine en forma completa, acorde con los datos suministrados, los campos principales de los encabezados de los segmentos que se generarán.

### 5

Datagrama IP de 832 bytes -> 812 bytes de segmento de la capa de transporte + 20 bytes de cabecera IP

El primer fragmento que se enviará llevará los primeros **496** bytes de datos (mayor múltiplo de 8 posible)

El segundo y último fragmento llevará el resto de los datos (316 bytes de datos)

|Fragmento  |Bytes                        |Largo  |Offset         |Indicador  |
|-----------|-----------------------------|-------|---------------|-----------|
|1          |496 de datos + 20 de cabecera|516    |D = 0          |1          |
|2          |316 de datos + 20 de cabecera|336    |D = 496/8 = 62 |0          |

Al encapsularse los mensajes de RIP en datagramas UDP: 

        a. RIP no sufre de problemas de la fragmentación de datagramas IP
        b. RIP no sufre los problemas de descartado de paquetes por congestión en encaminadores routers.

### 14.

a) Incorrecto. El tamaño del datagrama RIP encapsulado viaja a través de UDP, por lo tanto puede sufrir fragmentación.

b) Incorrecto. RIP se transmite a través de UDP, por lo que para la red es una paquete más (que se transmite por broadcast), por lo que puede llegar a ser descartado por congestión.

c) Correcto. Según la que describe el protocolo se dice que las peticiones se reciben al puerto 520.

d) Incorrecto, el número de protocolo va a ser el de UDP. Luego, en el router, se especifica que se revisa el número de protocolo del paquete RIP, ya desencapsulado.
